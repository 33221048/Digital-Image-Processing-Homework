In signal processing, the rectangular function, often denoted as $\left( \text{rect}(t) \right)$, is a function that takes the value 1 within a specific interval and 0 elsewhere. It's defined as:

$$
\text{rect}(t) = \begin{cases}
1 & \text{if } |t| \leq \frac{1}{2} \\
0 & \text{otherwise}
\end{cases}
$$

This function is also known as the boxcar function or the window function in some contexts. It's widely used in signal processing for various purposes, including filtering, windowing, and pulse shaping.

The function $\left( \text{rect}(\frac{t}{2}) \right)$ means the same as $\left( \text{rect}(t) \right)$, but it has a width of 1 unit centered around the origin. This function is often used in signal processing to represent a unit-width pulse centered at the origin.

```python
##########Your code here###########
def rect(t):
    """
    Rectangular function that returns 1 for |t| <= 1 and 0 otherwise.

    Parameters:
    t : float or np.ndarray
        The input time variable(s).

    Returns:
    float or np.ndarray
        The value of rect(t/2) at the given input.
    """
    # Calculate the value of rect(t/2)
    return np.where(np.abs(t) <= 1, 1, 0)

# Define time parameters
t_start = -1
t_end = 2
num_steps = 300
t = np.linspace(t_start, t_end, num_steps)

# Generate the rect function values
rect_values = rect(t)

# Plot rect function
plt.figure(figsize=(10, 5))
plt.stem(t, rect_values, label='rect(t/2)')
plt.title('rect(t/2)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.show()

# Compute Fourier Transform
fourier_transform = np.fft.fft(rect_values)
frequencies = np.fft.fftfreq(len(t), t[1] - t[0])

# Plot Fourier Transform
plt.figure(figsize=(10, 5))
plt.stem(frequencies, np.abs(fourier_transform))
plt.title('Fourier Transform of rect(t/2)')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.grid(True)
plt.show()
###################################
```

   - Increasing the number of time steps typically results in a finer resolution in the time domain. This means that the signal is sampled more densely, providing more detail about its shape and behavior over time.

2. **Time Interval**:
   - Changing the time interval affects the range over which the signal is sampled. Extending the time interval widens the range over which the signal is observed, providing more context about its behavior over a longer period.
   - In the frequency domain, a wider time interval can impact the range of frequencies captured in the FFT result. It may result in a wider range of frequencies being represented, allowing for better resolution of both low and high-frequency components.

In summary, changing these values can impact the resolution, detail, and frequency range represented in both the time and frequency domains of the FFT result. Increasing the number of time steps and widening the time interval generally provide more detailed and accurate representations of the signal and its frequency content.

The Fourier Transform, particularly in two dimensions (2D), is a powerful tool in signal processing, allowing us to analyze the frequency content of signals and images. Let's address your questions one by one:

### a) What is each axis in 2D-Fourier space representing?

In 2D-Fourier space (often visualized after applying the 2D Fourier Transform to a spatial domain signal or image), each axis represents the frequency components of the original signal/image along the corresponding spatial axis:
- The **x-axis** in 2D-Fourier space represents the frequency content along the horizontal axis of the original 2D signal or image. It shows how fast the signal varies along the x-direction.
- The **y-axis** represents the frequency content along the vertical axis of the original 2D signal or image, indicating the variation speed of the signal along the y-direction.

The origin $(0,0)$ in 2D-Fourier space denotes the zero frequency, or DC component, which essentially corresponds to the average or baseline intensity of the image. Frequencies increase in magnitude as they move away from the origin along either axis, representing higher spatial frequencies (finer details) in the signal/image.

### b) What would happen if the values are not symmetric around $(0,0)$ in the 2D-Fourier domain?

Fourier Transforms of real-valued signals or images are usually symmetric around the origin $(0,0)$ due to the Hermitian property of the Fourier Transform of real functions. If the values in the 2D-Fourier domain are not symmetric around $(0,0)$, it suggests a couple of possibilities:
- The original signal/image might be complex-valued, where the symmetry typically observed with real-valued signals does not strictly apply. In this case, the asymmetry can carry additional information about the phase of the complex-valued components.
- Asymmetry can also be introduced by processing artifacts or by specific manipulations in the Fourier domain (e.g., filtering, shifting), which may alter the natural symmetric properties of the Fourier Transform of real-valued signals.

### c) Given the 2D signal $ \sin(3\pi x + 4\pi y) $, find the maximum sampling intervals and the Nyquist rates.

The given signal is $ \sin(3\pi x + 4\pi y) $. To avoid aliasing and accurately sample this signal, we need to follow the Nyquist theorem, which states that the sampling rate should be at least twice the highest frequency present in the signal.

The frequencies in the signal can be identified from the coefficients of $x$ and $y$ inside the sine function:
- The frequency in $x$ direction: $3\pi$ radians per unit distance, equivalent to $1.5$ cycles per unit distance.
- The frequency in $y$ direction: $4\pi$ radians per unit distance, equivalent to $2$ cycles per unit distance.

The highest frequency present in the signal is along the $y$-direction with $2$ cycles per unit distance, so:
- The Nyquist rate for $y$ is $4$ samples per unit distance (twice the highest frequency).
- The Nyquist rate for $x$ is $3$ samples per unit distance (twice the frequency in $x$).

The maximum sampling intervals, which are the inverses of the Nyquist rates, would be:
- For $x$: $\frac{1}{3}$ units per sample.
- For $y$: $\frac{1}{4}$ units per sample.

This ensures that the signal is sampled adequately to capture all the information without introducing aliasing.

```python
def fft_image(img):
    # Compute the Fourier Transform
    f_img = np.fft.fft2(img)
    # Shift the zero frequency component to the center
    f_img_shifted = np.fft.fftshift(f_img)
    return f_img_shifted

def ifft_image(f_img):
    # Shift the zero frequency component back to the corner
    f_img_unshifted = np.fft.ifftshift(f_img)
    # Compute the Inverse Fourier Transform
    img_back = np.fft.ifft2(f_img_unshifted)
    # Take only the abs part (discarding imaginary part)
    img_back = np.abs(img_back)
    return img_back

# Demonstrate you code's functionality
##########Your code here###########
# Load the provided fft file
fft_image_data = np.load("fft_file.npy")

# Display the FFT of the image
plt.subplot(1, 2, 1)
plt.imshow(np.log(1 + np.abs(fft_image_data)), cmap='gray')
plt.title('FFT of Image')
plt.axis('off')

# Reconstruct the image from FFT
reconstructed_img = ifft_image(fft_image_data)

# Display the reconstructed image
plt.subplot(1, 2, 2)
plt.imshow(reconstructed_img, cmap='gray')
plt.title('Reconstructed Image')
plt.axis('off')

plt.show()

Image.fromarray(((1 - reconstructed_img) * 255).round().astype(np.uint8)).save("shiba_inu.jpg")
###################################
```

This is a necessary step in converting analog signals to digital form, which is required for digital signal processing, storage, and transmission. Quantization inevitably introduces quantization error, but the goal is to minimize this error as much as possible. [1][2]

In summary, quantization is a fundamental process in converting analog signals to digital form, and the Lloyd-Max quantizer is a technique for optimizing the quantization levels to minimize the mean squared quantization error.

```python
def quantizer(image, decision_levels, reconstruction_levels):
  """
  This function performs uniform quantization on a grayscale image.

  Args:
      image: A NumPy array representing the grayscale image.
      decision_levels: A list containing the decision levels for quantization.
      reconstruction_levels: A list containing the reconstruction levels for each quantization bin.

  Returns:
      A NumPy array representing the quantized grayscale image.
  """
  # Flatten the image to a 1D array for easier processing
  flat_image = image.flatten()

  # Quantize each pixel value based on decision levels
  quantized_values = np.digitize(flat_image, decision_levels) - 1

  # Assign reconstruction levels based on the quantized values
  quantized_image = reconstruction_levels[quantized_values]

  # Reshape the quantized image back to its original shape
  return quantized_image.reshape(image.shape)
```

```python
def lloyd_max_quantizer(image, num_levels):
    # Flatten the image to simplify the computation
    flattened_image = image.flatten()

    # Initial guess for reconstruction values (quantization levels) - equally spaced
    reconstruction_values = np.linspace(flattened_image.min(), flattened_image.max(), num_levels)

    # Placeholder for decision levels
    decision_levels = np.zeros(num_levels + 1)
    decision_levels[-1] = np.inf  # Ensuring the last decision level captures all higher values

    for _ in tqdm(range(100)):  # Iterate a fixed number of times for simplicity; consider a convergence condition for a full implementation
        # Update decision levels based on current reconstruction values

        prev_decision_levels = decision_levels.copy()
        for i in range(1, num_levels):
            decision_levels[i] = (reconstruction_values[i - 1] + reconstruction_values[i]) / 2

        # Update reconstruction values based on current decision levels
        for i in range(num_levels):
            # Mask the values between current decision levels
            mask = (flattened_image >= decision_levels[i]) & (flattened_image < decision_levels[i + 1])
            if np.any(mask):
                reconstruction_values[i] = np.mean(flattened_image[mask])  # Update with mean of values in the range

        # Check for convergence
        if np.allclose(prev_decision_levels, decision_levels):
            break

    # Adjust the first and last decision levels to match the image intensity range
    decision_levels[0] = flattened_image.min()
    decision_levels[-2] = flattened_image.max()  # -2 because the last value is np.inf
    print(f'{decision_levels[:-1]=}')
    print(f'{reconstruction_values=}')

    return decision_levels[:-1], reconstruction_values
```

```python
def uniform_quantizer(image, num_levels):
    # Find the min and max intensity values in the image
    min_val = image.min()
    max_val = image.max()

    # Calculate the step size for the given number of levels
    step_size = (max_val - min_val) / num_levels

    # Calculate reconstruction values (quantization levels)
    # These are the mid-points of each quantization interval
    reconstruction_values = np.linspace(min_val + step_size / 2, max_val - step_size / 2, num_levels)

    # Calculate decision levels
    # Decision levels are the boundaries at which the signal is quantized to the next level
    decision_levels = np.linspace(min_val, max_val, num_levels)
    print(f'{decision_levels=}')
    print(f'{reconstruction_values=}')

    return decision_levels, reconstruction_values
```

```python
decision_levels, reconstruction_values = uniform_quantizer(img, 5)
uniform_quantized_img = quantizer(img, decision_levels, reconstruction_values)
uniform_mse = calculate_mse(img, uniform_quantized_img)

plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(quantized_img, cmap='gray')
plt.title(f"Uniform Image {len(decision_levels)} bins ({uniform_mse:.2f})")
plt.axis("off")

plt.tight_layout()
plt.show()
```
